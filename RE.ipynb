## Huawei info

>Python regular expressions are special sequences of characters and they enable easy check on whether a string matches a mode.

Регулярные выражения Python - это специальные последовательности символов, которые позволяют легко проверить, соответствует ли строка определённым правилам.
***
>Python of **version 1.5** and later has the new re module, which provides a Perl-style regular expression mode.
The re module enables Python to have all regular expression functions.

Python **версии 1.5** и выше имеет новый модуль re, который обеспечивает режим регулярных выражений в стиле Perl.
Модуль re позволяет Python иметь все функции регулярных выражений. (Вопрос)
